```python
#Import the required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

#Download the webpage at the url
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data = requests.get(url).text

#Create a soup object
soup = BeautifulSoup(data,'html.parser')

#Scrape the table: Scrape the Language name and annual average salary
table = soup.find('table')

#Get a dataframe
salary_data = pd.DataFrame(columns=["No.","Language","Created By","Average Annual Salary","Learning Difficulty"])

for row in table.find_all("tr"):
    col = row.find_all("td")
    if col:
        no = col[0].text.strip()
        language = col[1].text.strip()
        created_by = col[2].text.strip()
        avg_salary = col[3].text.strip()
        learn = col[4].text.strip()

        # Create a temporary DataFrame for the new row
        new_row = pd.DataFrame([{"No.": no,"Language": language,"Created By": created_by,"Average Annual Salary": avg_salary,"Learning Difficulty": learn}])

        # Use concat
        salary_data = pd.concat([salary_data, new_row], ignore_index=True)

salary_data
```

```python
#Save the scrapped data into a file named popular-languages.csv
salary_data.to_csv('popular-languages.csv')
```
